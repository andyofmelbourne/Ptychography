```python
def translate(array, coord):
    x, y     = bg.make_xy(array.shape)
    exp      = np.exp(-2.0J * np.pi * (coord[1] * x / float(array.shape[1]) + coord[0] * y / float(array.shape[0])) )
    return bg.ifft2(bg.fft2(array) * exp)

def makeExits(sample, probe, coords):
    exits = []
    for coord in coords:
        exits.append(sample * translate(probe, coord))
    return np.array(exits)

def makeDiffs(sample, probe, coords):
    diffs = np.abs(bg.fft2(makeExits(sample, probe, coords)))**2
    return diffs

diffs = makeDiffs(sample_0, probe_0, coords_0)
```

```python
def emod(diffs, psis):
    psis = bg.fft2(psis)
    error = np.sum( (np.sqrt(diffs) - np.abs(psis))**2 )
    return error
```

```python
class Cgls(object):
    """A class for minimising a non-linear function f(x) with respect to x using the non-linear cgls algorithm. 

    x is a list of numpy arrays."""
        
    def __init__(self, x_0, fgrad, fgrad_dot, fgrad2_dot):
        """
        x_0 is the initial state vector (list of numpy arrays)
        fgrad is a function for calculating the gradient (= fgrad(x))
        fgrad_dot calculates fgrad . x' (=fgrad_dot(x', x))
        fgrad_dot calculates x' . fgrad2 . x' (=fgrad2_dot(x', x))
        """
        self.x          = list(x_0)
        self.fgrad      = fgrad
        self.fgrad_dot  = fgrad_dot
        self.fgrad2_dot = fgrad2_dot
        self.iterations = 0
        self.error_residual = []
    
    def aT_dot_b(self, a, b):
        """Calculate np.sum(np.conj(a) * b) but for lists"""
        tot = 0.0
        for aa, bb in zip(a, b):
            tot += np.sum(np.conj(aa) * bb)
        return tot
    
    def calc(self, iterations = 10):
        """Iteratively solve the non-linear equations using the conjugate gradient method.
        
        All of the vectors are 'selfed' so that the iterations may continue 
        when called again."""
        if self.iterations == 0 :
            self.r     = list(( -f for f in self.fgrad(self.x) ))
            self.d     = list(self.r)
                                    
        i = 0
        for i in range(iterations):
            # line search f(x_i + alpha d_i)_min alpha
            # alpha = -fgrad_dot(d_i, x_i) / fgrad2_dot(d_i, x_i)
            alpha = -self.fgrad_dot(self.d, self.x) / self.fgrad2_dot(self.d, self.x)
            #
            # increment
            self.x    = list(( x + alpha * d for x,d in zip(self.x, self.d)))
            #
            self.rr   = list(( -f for f in self.fgrad(self.x) ))
            #
            dr  = list(( rr - r for rr,r in zip(self.rr, self.r)))
            res = self.aT_dot_b(self.r, self.r)
            if res <= 1.0e-10:
                break
            self.beta = np.abs(self.aT_dot_b(self.rr, dr) / res)
            self.beta = max(self.beta, 0.0)
            #
            self.d = list(( rr + self.beta*d for rr,d in zip(self.rr, self.d)))
            #
            self.r = list(self.rr)
            #
            if self.iterations % 4 == 0:
                self.r     = list(( -f for f in self.fgrad(self.x) ))
                self.d     = list(self.r)
            
        self.iterations += i + 1
        return self.x
        
# Test cgls
#f          = lambda x: (x[0]-1)**4 + 1
#x_0        = [100.0]
#fgrad      = lambda x: [4*(x[0]-1)**3]
#fgrad_dot  = lambda xd, x: xd[0] * 4*(x[0]-1)**3
#fgrad2_dot = lambda xd, x: xd[0]**2 * 12*(x[0]-1)**2
    
#prob = Cgls(x_0, fgrad, fgrad_dot, fgrad2_dot)

#error = []
#for i in range(100):
#    prob.calc(1)
#    error.append(np.abs(prob.x[0] - 1)**2)

#plot(np.log(np.array(error)), label='error Vs iters')
#legend()
```

```python
class Pcgls(object):
    def __init__(self, x_0, diffs):
        self.x_0   = x_0
        self.diffs = diffs
        self.cgls = Cgls(x_0, self.fgrad, self.fgrad_dot, self.fgrad2_dot)
    
    def f(self, x):
        return self.emod(self.diffs, self.makeExits(x))
    
    def fgrad(self, x):
        return self.grad(x)
    
    def fgrad_dot(self, x_d, x):
        return self.emod_grad(x_d, x)
        
    def fgrad2_dot(self, x_d, x):
        return self.emod_grad2(x_d, x)

    def Pmod(self, psis):
        psis = bg.fft2(psis)
        psis = np.sqrt(self.diffs) * psis / np.abs(psis)
        return bg.ifft2(psis)
    
    def Pmod_hat(self, psis):
        psis2 = np.sqrt(self.diffs) * psis / np.abs(psis)
        return psis2
    
    def translate(self, array, coord):
        x, y     = bg.make_xy(array.shape)
        exp      = np.exp(-2.0J * np.pi * (coord[1] * x / float(array.shape[1]) + coord[0] * y / float(array.shape[0])) )
        return bg.ifft2(bg.fft2(array) * exp)

    def makeProbes(self, probe, coords):
        probes = []
        for coord in coords:
            probes.append(self.translate(probe, coord))
        return np.array(probes)
    
    def makeExits(self, x):
        exits = x[0] * self.makeProbes(x[1], x[2])
        return exits

    def makeDiffs(self, x):
        diffs = np.abs(bg.fft2(self.makeExits(x)))**2
        return diffs
    
    def emod(self, diffs, psis):
        psis  = bg.fft2(psis)
        error = np.sum( (np.sqrt(diffs) - np.abs(psis))**2 )
        return error
    
    def grad(self, x):
        x_out  = list(x)
        probes = self.makeProbes(x[1], x[2])
        exits  = x[0] * probes
        exits  = exits - self.Pmod(exits)
        # grad T
        x_out[0] = 2 * np.sum(np.conj(probes) * exits, axis=0)
        # grad P
        Ts = self.makeProbes(x[0], -x[2])
        x_out[1] = 2 * np.sum(np.conj(Ts) * (x[1] * Ts - self.Pmod(x[1] * Ts)), axis=0)
        # 
        # grad R
        exits_z    = x[0] * self.makeProbes_z(x[1], x[2])
        exits_zbar = x[0] * self.makeProbes_zbar(x[1], x[2])
        exits      = np.conj(exits_z) * exits + np.conj(exits) * exits_zbar
        arrayout   = np.sum(exits, axis=(1,2))
        arrayout2  = np.zeros_like(x[2])
        arrayout2[:, 0] = np.real(arrayout)
        arrayout2[:, 1] = np.imag(arrayout)
        x_out[2]   = 2 * arrayout2
        #
        # testing
        x_out[0].fill(0.0)
        x_out[2].fill(0.0)
        a = x_out[1][1,1]
        x_out[1].fill(0.0)
        x_out[1][1,1] = a
        return x_out
    
    def emod_grad(self, x_d, x):
        """x_d is the search direction and x the current state vector. x is a list of numpy arrays x = [T, probe, coords]"""
        exits    = self.makeExits(x)
        #exits_Td = self.makeExits([x_d[0], x[1]  , x[2]])
        exits_Pd = self.makeExits([x[0]  , x_d[1], x[2]])
        #exits_Rd = self.makeExits_grad([x[0]  , x[1]  , x[2]], x_d[2])
        #exits = np.conj(exits - self.Pmod(exits)) * (exits_Td + exits_Pd + exits_Rd)
        # testing
        exits = np.conj(exits - self.Pmod(exits)) * (exits_Pd)
        return 2 * np.sum( np.real(exits) ) 
    
    def emod_grad2(self, x_d, x):
        """x_d is the search direction and x the current state vector. x is a list of numpy arrays x = [T, probe, coords]"""
        exits     = bg.fft2(self.makeExits(x))
        exits_Td  = bg.fft2(self.makeExits([x_d[0], x[1]  , x[2]]))
        exits_Pd  = bg.fft2(self.makeExits([x[0]  , x_d[1], x[2]]))
        exits_TPd = bg.fft2(self.makeExits([x_d[0], x_d[1]  , x[2]]))
        exits_TRd = bg.fft2(self.makeExits_grad([x_d[0] , x[1], x[2]], x_d[2]))
        exits_Rd  = bg.fft2(self.makeExits_grad([x[0] , x[1], x[2]], x_d[2]))
        exits_RRd = bg.fft2(self.makeExits_grad2([x[0], x[1], x[2]], x_d[2]))
        #
        temp = np.sqrt(self.diffs) / np.abs(exits)**3
        # testing
        arrayout  = 2 * np.abs(exits_Pd)**2 - 2 * temp * (np.imag(exits * np.conj(exits_Pd))**2 )
        # T . H . T + P . H . P
        #arrayout  = 2 * np.abs(exits_Td)**2 + 2 * np.abs(exits_Pd)**2 - 2 * temp * (np.imag(exits * np.conj(exits_Td))**2 + np.imag(exits * np.conj(exits_Pd))**2)
        # R . H . R 
        #arrayout += 2 * (np.conj(exits_RRd) * (exits - self.Pmod_hat(exits)) + np.abs(exits_Rd)**2 * (1 - np.sqrt(self.diffs) / np.abs(exits)) + temp * np.real(np.conj(exits) * exits_Rd)**2)
        # 2 x T . H . P
        #arrayout += 4 * (np.conj(exits_Td) * exits_Pd + np.conj(exits_TPd) * (exits - self.Pmod_hat(exits)) + temp * np.imag(np.conj(exits_Td) * exits) * np.imag(np.conj(exits_Pd) * exits))
        # 2 x T . H . R
        #arrayout += 4 * (np.conj(exits_TRd) * exits_Rd + np.conj(exits_TRd) * (exits - self.Pmod_hat(exits)) - temp * np.real(np.conj(exits_Td) * exits) * np.real(np.conj(exits_Rd) * exits))
        # 2 x P . H . R
        #exits     = bg.fft2(self.makeExits(x))
        #exits_TRd = bg.fft2(self.makeExits_grad([x_d[1] , x[0], x[2]], x_d[2]))
        #exits_Td  = bg.fft2(self.makeExits([x_d[1], x[0]  , x[2]]))
        #exits_Rd  = bg.fft2(self.makeExits_grad([x[1] , x[0], x[2]], x_d[2]))
        #arrayout -= 4 * (np.conj(exits_TRd) * exits_Rd + np.conj(exits_TRd) * (exits - self.Pmod_hat(exits)) - temp * np.real(np.conj(exits_Td) * exits) * np.real(np.conj(exits_Rd) * exits))
        return np.sum( np.real(arrayout) ) 
    
    def makeProbes_z(self, probe, coords):
        exits = []
        for coord in coords:
            exits.append(self.probe_grad_z(probe, coord))
        return np.array(exits)
    
    def makeProbes_zbar(self, probe, coords):
        exits = []
        for coord in coords:
            exits.append(self.probe_grad_zbar(probe, coord))
        return np.array(exits)
    
    def probe_grad_z(self, probe, coord):
        x, y     = bg.make_xy(probe.shape)
        exp      = np.exp(-2.0J * np.pi * (coord[1] * x / float(probe.shape[1]) + coord[0] * y / float(probe.shape[0])) )
        array    = -1.0J * np.pi * (y / float(probe.shape[0]) -  1J*x / float(probe.shape[1]))
        return bg.ifft2(bg.fft2(probe) * exp * array)
    
    def probe_grad_zbar(self, probe, coord):
        x, y     = bg.make_xy(probe.shape)
        exp      = np.exp(-2.0J * np.pi * (coord[1] * x / float(probe.shape[1]) + coord[0] * y / float(probe.shape[0])) )
        array    = -1.0J * np.pi * (y / float(probe.shape[0]) + 1J*x / float(probe.shape[1]))
        return bg.ifft2(bg.fft2(probe) * exp * array)
    
    def probe_trans_grad(self, probe, coord, coord2):
        """I will also apply a basic Gaussian filter to this. Not sure what the optimal solution is.
        coord is the shift coordinate while coord2 is the multiplicative component.
        """
        std      = [probe.shape[0]/3.0, probe.shape[1]/3.0]
        #std      = [1.0, 1.0]
        x, y     = bg.make_xy(probe.shape)
        exp      = np.exp(-2.0J * np.pi * (coord[1] * x / float(probe.shape[1]) + coord[0] * y / float(probe.shape[0])) )# - 0.5 * ( (x / std[1])**2 + (y / std[0])**2 ) )
        array    = -2.0J * np.pi * (coord2[0] * y / float(probe.shape[0]) + coord2[1] * x / float(probe.shape[1]))
        return bg.ifft2(bg.fft2(probe) * exp * array)
    
    def probe_trans_grad2(self, probe, coord, coord2):
        #std      = [1.0, 1.0]
        std      = [probe.shape[0]/3.0, probe.shape[1]/3.0]
        x, y     = bg.make_xy(probe.shape)
        exp      = np.exp(-2.0J * np.pi * (coord[1] * x / float(probe.shape[1]) + coord[0] * y / float(probe.shape[0])) )#- 0.5 * ( (x / std[1])**2 + (y / std[0])**2 ) )
        array    = (-2.0J * np.pi * (coord2[0] * y / float(probe.shape[0]) + coord2[1] * x / float(probe.shape[1])))**2
        return bg.ifft2(bg.fft2(probe) * exp * array)
    
    def makeExits_grad(self, x, coords2):
        exits = []
        for i in range(len(coords2)):
            exits.append(probe_trans_grad(x[1], x[2][i], coords2[i]))
        return np.array(exits) * x[0]

    def makeExits_grad2(self, x, coords2):
        exits = []
        for i in range(len(coords2)):
            exits.append(probe_trans_grad2(x[1], x[2][i], coords2[i]))
        return np.array(exits) * x[0]
        
```
