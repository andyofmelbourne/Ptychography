# Make Diffraction data

```python
def translate(array, coord):
    x, y     = bg.make_xy(array.shape)
    exp      = np.exp(-2.0J * np.pi * (coord[1] * x / float(array.shape[1]) + coord[0] * y / float(array.shape[0])) )
    return bg.ifft2(bg.fft2(array) * exp)

def makeExits(sample, probe, coords):
    exits = []
    for coord in coords:
        exits.append(sample * translate(probe, coord))
    return np.array(exits)

def makeDiffs(sample, probe, coords):
    diffs = np.abs(bg.fft2(makeExits(sample, probe, coords)))**2
    return diffs

diffs = makeDiffs(sample_0, probe_0, coords_0)
```

```python
def emod(diffs, psis):
    psis = bg.fft2(psis)
    error = np.sum( (np.sqrt(diffs) - np.abs(psis))**2 )
    return error
```

```python
class cgls(object):
    """A class for minimising a non-linear function f(x) with respect to x using the non-linear cgls algorithm. 

    x is a list of numpy arrays."""def emod(diffs, psis):
    psis = bg.fft2(psis)
    error = np.sum( (np.sqrt(diffs) - np.abs(psis))**2 )
    return error
    
    def __init__(self, x_0, fgrad, fgrad_dot, fgrad2_dot):
        """
        x_0 is the initial state vector (list of numpy arrays)
        fgrad is a function for calculating the gradient (= fgrad(x))
        fgrad_dot calculates fgrad . x' (=fgrad_dot(x', x))
        fgrad_dot calculates x' . fgrad2 . x' (=fgrad2_dot(x', x))
        """
        self.x          = list(x_0)
        self.fgrad      = fgrad
        self.fgrad_dot  = fgrad_dot
        self.fgrad2_dot = fgrad2_dot
        self.iterations = 0
        self.error_residual = []
    
    def aT_dot_b(self, a, b):
        """Calculate np.sum(np.conj(a) * b) but for lists"""
        tot = 0.0
        for aa, bb in zip(a, b):
            tot += np.sum(np.conj(a) * b)
        return tot
    
    def calc(self, iterations = 10):
        """Iteratively solve the non-linear equations using the conjugate gradient method.
        
        All of the vectors are 'selfed' so that the iterations may continue 
        when called again."""
        if self.iterations == 0 :
            self.r     = list(( -f for f in self.fgrad(self.x) ))
            self.d     = list(self.r)
                    
        for i in range(iterations):
            # line search f(x_i + alpha d_i)_min alpha
            # alpha = -fgrad_dot(d_i, x_i) / fgrad2_dot(d_i, x_i)
            alpha = -fgrad_dot(self.d, self.x) / fgrad2_dot(self.d, self.x)
            #
            # increment
            self.x    = list(( x + alpha * d for x,d in zip(self.x, self.d)))
            #
            self.rr   = list(( -f for f in self.fgrad(self.x) ))
            #
            dr  = list(( rr - r for rr,r in zip(self.rr, self.r)))
            res = self.aT_dot_b(self.r, self.r)
            if res <= 1.0e-10:
                break
            beta = self.aT_dot_b(self.rr, dr) / res
            beta = max(beta, 0.0)
            #
            self.d = list(( rr + beta*d for rr,d in zip(self.rr, self.d)))
            #
            self.r = list(self.rr)
            
        self.iterations += i + 1
        return self.x
        
# Test cgls
f          = lambda x: (x[0]-1)**4 + 1
x_0        = [100.0]
fgrad      = lambda x: [4*(x[0]-1)**3]
fgrad_dot  = lambda xd, x: xd[0] * 4*(x[0]-1)**3
fgrad2_dot = lambda xd, x: xd[0]**2 * 12*(x[0]-1)**2
    
prob = cgls(x_0, fgrad, fgrad_dot, fgrad2_dot)

error = []
for i in range(100):
    prob.calc(1)
    error.append(np.abs(prob.x[0] - 1)**2)

plot(np.log(np.array(error)))
```
