```python
shape = (1024,)

T_0      = np.random.random(shape) + 1J*np.random.random(shape) 
probe_0  = np.zeros_like(T_0)
probe_0[shape[0]/4 : shape[0] * 3/4] = 1.0
coords_0 = [1, 10]
```

# The diffraction data

```python
def translate(array, coord):
    x        = np.arange(array.shape[-1])
    x        = np.roll(x, -(array.shape[-1] // 2 - 1), 0)
    x        = x - x[0]
    exp      = np.exp(-2.0J * np.pi * (coord * x / float(array.shape[0])) )
    return bg.ifftn_1d(bg.fftn_1d(array) * exp)

def makeProbes(probe, coords):
    exits = []
    for coord in coords:
        exits.append(translate(probe, coord))
    return np.array(exits)

def makeExits(sample, probe, coords):
    exits = []
    for coord in coords:
        exits.append(sample * translate(probe, coord))
    return np.array(exits)

def makeDiffs(sample, probe, coords):
    diffs = np.abs(bg.fftn_1d(makeExits(sample, probe, coords)))**2
    return diffs

diffs = makeDiffs(T_0, probe_0, coords_0)
```

# The metric

```python
def emod(diffs, exits):
    exits = bg.fftn_1d(exits)
    error = np.sum( (np.sqrt(diffs) - np.abs(exits))**2 )
    return error

def fmod(diffs, T, probe, coords, precon = None):
    if np.all(precon != None) :
        diffs2 = makeDiffs(T, probe, coords)
        return np.sum((precon * (np.sqrt(diffs2) - np.sqrt(diffs)))**2) / np.sum(precon**2 * diffs)
    return np.sum((np.sqrt(makeDiffs(T, probe, coords)) - np.sqrt(diffs))**2) / np.sum(diffs)

def makeExits_grad(sample, probe, coords, coords2):
    exits = []
    for i in range(len(coords)):
        exits.append(sample * probe_trans_grad(probe, coords[i], coords2[i]))
    return np.array(exits)

def probe_trans_grad(probe, coord, coord2):
    Nx       = probe.shape[0]
    x        = np.arange(diffs.shape[-1])
    x        = np.roll(x, -(diffs.shape[-1] // 2 - 1), 0)
    x        = x - x[0]
    exp      = np.exp(-2.0J * np.pi * coord * x / float(Nx))
    array    = -2.0J * np.pi * (coord2 * x / float(Nx))
    return bg.ifftn_1d(bg.fftn_1d(probe) * exp * array)

def emod_grad_coords(coords_d, T, probe, coords, diffs, precon = None):
    exits_d = bg.fftn_1d(makeExits_grad(T, probe, coords, coords_d))
    exits   = bg.fftn_1d(makeExits(T, probe, coords))
    exits   = np.conj(exits_d) * (exits - Pmod_hat(diffs, exits))
    if np.all(precon != None):
        exits = exits * np.sqrt(precon)
    return 2 * np.sum(np.real(exits) )

def Pmod(diffs, psis):
    psis = bg.fftn_1d(psis)
    psis = np.sqrt(diffs) * psis / np.abs(psis)
    return bg.ifftn_1d(psis)

def Pmod_hat(diffs, psis):
    a = 1.0 / np.abs(psis)
    if np.any(a == None) or np.any(a == np.inf) :
        print 'divide by zero error'
    psis2 = np.sqrt(diffs) * psis * a
    return psis2

def grad_coords(T, probe, coords):
    exits      = makeExits(T, probe, coords)
    exits_z    = T * makeProbes_z(probe, coords)
    exits      = (exits - Pmod(diffs, exits))
    exits      = np.conj(exits_z) * exits
    arrayout   = np.sum(exits, axis=(-1))
    arrayout2  = np.zeros_like(coords)
    arrayout2  = np.real(arrayout)
    return 2 * arrayout2

def makeProbes_z(probe, coords):
    exits = []
    for coord in coords:
        exits.append(probe_grad_z(probe, coord))
    return np.array(exits)

def probe_grad_z(probe, coord):
    x        = np.arange(probe.shape[-1])
    x        = np.roll(x, -(probe.shape[-1] // 2 - 1), 0)
    x        = x - x[0]
    exp      = np.exp(-2.0J * np.pi * coord * x / float(probe.shape[-1]))
    array    = -1.0J * np.pi * (x / float(probe.shape[-1]))
    return bg.ifftn_1d(bg.fftn_1d(probe) * exp * array)
```

    \end{align}
</math>
